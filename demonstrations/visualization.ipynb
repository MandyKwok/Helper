```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# generate freq plot for dates
def plot_date(series):
    """
    Input: a pd series or a list of datetime objects
    """
    if type(series) == 'pandas.core.series.Series':
        series = series.dropna()
    
    date_list = list(series)
    date_list = [x for x in date_list if str(x) != 'nan']
    value_counts = [date_list.count(x) for x in date_list]

    dates = mdates.date2num(date_list)
    
    # set the size
    fig = plt.gcf()
    fig.set_size_inches(18.5, 10.5)
    
    # plot
    plt.plot_date(dates, value_counts) 
```

```python
# generate a list of random dates

from random import randrange
from datetime import timedelta, datetime

start = datetime(2020,1,1,0,0,0).astimezone().replace(microsecond=0)
end = datetime(2020,1,31,23,59,59).astimezone().replace(microsecond=0)

def generate_N_random_date(N, start, end):
    """
    This function will return N random datetime objects between two datetime 
    objects.
    """
    delta = end - start
    int_delta = (delta.days * 24 * 60 * 60) + delta.seconds
    result = []
    for i in range(N):
        random_second = randrange(int_delta)
        datetime_object = datetime.date(start + timedelta(seconds = random_second))
        result.append(datetime_object)
    return result

random_dates = generate_N_random_date(N = 1000, start = start, end = end)
```

```python
import random
import datetime

def generate_N_random_date(N, start, end):
    """
    This function will return N random datetime objects between two datetime 
    objects.
    """
    delta = end - start
    int_delta = (delta.days * 24 * 60 * 60) + delta.seconds
    
    result = []
    for i in range(N):
        random_second = random.randrange(int_delta)
        datetime_object = datetime.datetime.date(start + datetime.timedelta(seconds = random_second))
        result.append(datetime_object)
        
    return result

random_dates = generate_N_random_date(N = 1000, start = start, end = end)
```

```python
plot_date(random_dates)
```
